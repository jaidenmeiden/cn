```python

def sublist(lst1, lst2):
    return set(lst1) <= set(lst2)

def sublist2(lst1, lst2):
   ls1 = [element for element in lst1 if element in lst2]
   ls2 = [element for element in lst2 if element in lst1]
   return ls1 == ls2

class simpleSS:
    
    def __init__(self, rules):
        if type(rules) is not list:
            self.__exist = False
            print('error, input for class has to be a one folded list')
        else:
            self.__exist = True
        if not rules:
            self.__empty = True
            print('error, input for class has to be non-empty list')
        else:
            self.__empty = False
        if self.__exist and not self.__empty:
            self.__input = list(rules)
            self.__input = self.__checkIfString(self.__input)
            self.__input = self.__lowerCase(self.__input)
            self.__input = self.__checkCorrect(self.__input)
            self.__entity = self.__input
            self.__entity = self.__convertEntity(self.__entity)
            self.__first = self.__fillFirst(self.__entity)
            self.__last = self.__fillLast(self.__entity)
            self.__loops = self.__fillLoops(self.__entity)
            self.__multiple = self.__fillMultiple(self.__entity)
            self.__mFlag = self.__isMultiple(self.__multiple)
            self.__gramma = self.__setGrammar(self.__first, self.__loops,
                                                self.__last, self.__mFlag)

    def __lowerCase(self, listvr):
        listTMP = []
        for w in listvr:
            listTMP.append(w.lower())
        return listTMP

    def __printConverted(self):
        for w in self.__entity:
            print(w)

    def __printGramma(self):
        return self.__gramma

    def __printEntity(self):
        return self.__entity

    def __printInput(self):
        return self.__input 

    def __printFirst(self):
        return self.__first

    def __printLast(self):
        return self.__last
    
    def __printLoops(self):
        return self.__loops

    def __printMultiple(self):
        return self.__multiple
    
    def __printIfB(self):
        return self.__mFlag

    def __convertEntity(self, entity):
        listTMP = []
        for w in entity:
            listTMP.append(self.__convertList(w))
        return listTMP

    def __convertList(self, listvr):
        listTMP = []
        before = listvr[0]
        count = 1
        for i in range(1, len(listvr)):
            if before == listvr[i]:
                count += 1
            else:
                listTMP.append(listvr[i-1])
                listTMP.append(count)
                count = 1
            before = listvr[i]
            if i == len(listvr)-1:
                listTMP.append(listvr[i])
                listTMP.append(count)
        return listTMP

    def __fillFirst(self, entity):
        listTMP = []
        for w in entity:
            if w[0] == 'a':
                listTMP.append(w[1])
        listTMP = list(dict.fromkeys(listTMP))
        listTMP.sort()
        return listTMP

    def __fillLast(self, entity):
        listTMP = []
        for w in entity:
            if w[-2] == 'a':
                listTMP.append(w[-1])
        listTMP = list(dict.fromkeys(listTMP))
        listTMP.sort()
        return listTMP

    def __fillLoops(self, entity):
        listTMP = []
        for w in entity:
            iterTMP = listTMP
            listTMP = iterTMP + self.__checkLoops(w)
        listTMP = list(dict.fromkeys(listTMP))
        listTMP.sort()
        return listTMP

    def __checkLoops(self, word):
        listTMP = []
        if word[0]=='b':
            k=2
        else:
            k=4
        for i in range(k,len(word)-2,4):
            if word[i] == 'a':
                listTMP.append(word[i+1])
        listTMP = list(dict.fromkeys(listTMP))
        return listTMP

    def __fillMultiple(self, entity):
        listTMP = []
        for w in entity:
            iterTMP = listTMP
            listTMP = iterTMP + self.__checkMultiple(w)
        listTMP = list(dict.fromkeys(listTMP))
        listTMP.sort()
        return listTMP

    def __checkMultiple(self, word):
        listTMP = []
        if word[0]=='b':
            k=0
        else:
            k=2
        for i in range(k,len(word),4):
            if word[i] == 'b':
                listTMP.append(word[i+1])
        listTMP = list(dict.fromkeys(listTMP))
        return listTMP

    def __isMultiple(self, bs):
        if bs: 
            flag=True if bs[-1]>1 else False
        else: flag = False
        return flag

    def printInfo(self):
        if self.__exist and not self.__empty:
            print('input')
            print(self.__printInput())
            print('converted input')
            #print(self.__printEntity())
            self.__printConverted()
            print('a in begining')
            print(self.__printFirst())
            print('a in middle')
            print(self.__printLoops())
            print('a in end')
            print(self.__printLast())
            print('b')
            print(self.__printMultiple())
            print('we have b loops')
            print(self.__printIfB())
            print('regular expression')
            print(self.__printGramma())
        else:
            print('class initialization error, print function')

    def __checkIfString(self, input):
        listTMP=[]
        for w in input:
            if type(w) is str:
                listTMP.append(w)
        return listTMP

    def __checkCorrect(self, input):
        listTMP=[]
        for w in input:
            if self.__checkCorrectWord(w):
                listTMP.append(w)
        return listTMP

    def __checkCorrectWord(self, word):
        allowed = {'a','b'}
        onlyA = {'a'}
        countA=0
        otherLetters=False
        for let in word:
            if let not in allowed:
                otherLetters=True
                break
            elif let in onlyA:
                countA+=1
        flag=countA<len(word) and not otherLetters
        return flag

    def __setGrammar(self, first, loops, last, flag, star=True):
        start = ''
        for s in first:
            start+=s*'a'
            start+='+'
        start=start[:-1]
        middle = ''
        if flag:
            middle+='b+'
        for s in loops:
            middle+=s*'a'
            middle+='b+'
        middle=middle[:-1]
        end = ''
        for s in last:
            end+=s*'a'
            end+='+'
        end=end[:-1]
        myGramma = '('+start+')'
        if myGramma:
            myGramma +='b'
        myGramma += '('+middle+')'
        if star: myGramma +='*'
        myGramma += '('+end+')'
        return myGramma

    def checkWord(self, word):
        if self.__exist and not self.__empty:
            if type(word) is str:
                check=True
                listTMP = [word]
                print(f'input "{word}"')
                w = list(listTMP)
                w = self.__checkIfString(w)
                w = self.__lowerCase(w)
                w = self.__checkCorrect(w)
                if w:

                    entity = w
                    entity = self.__convertEntity(entity)
                    first = self.__fillFirst(entity)
                    last = self.__fillLast(entity)
                    loops = self.__fillLoops(entity)
                    multiple = self.__fillMultiple(entity)
                    mFlag = self.__isMultiple(multiple)
                    gramma = self.__setGrammar(first, loops, last, mFlag, True)
                    print(f'check "{w[0]}"')
                    print(f'word is in {gramma}')
                    print('regular expression is')
                    print(self.__printGramma())
                    if not sublist(first,self.__first): check=False
                    if not sublist(last,self.__last): check=False
                    if not sublist(loops,self.__loops): check=False
                    if not sublist(multiple,self.__multiple): check=False
                else:
                    check=False
                if check:
                    print('this word can be obtained')
                    print()
                else:
                    print('this word cannot be obtained')
                    print()
            else:
                print('wrong input format, must be "a","b" string')
                print()
        else:
            print('class initialization error, check function')
```
